1. What is hypothesis testing in statistics ?

-> Hypothesis testing in statistics is a method used to validate assumptions or claims about data. It's a structured way to determine if the findings of a study provide evidence to support a specific theory relevant to a larger population.

->Null Hypothesis (H0), which is a statement of no effect or no difference.
-> An Alternative Hypothesis (H1), which is the opposite of the null hypothesis.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis ?

-> The null hypothesis is a statement that there's no effect or no difference in the population being studied. It's essentially the default assumption that there's no relationship between variables, and it's denoted by H0.

-> On the other hand, the alternative hypothesis (H1 or Ha) suggests that there is an effect or a difference in the population.

-> Key differences between null and alternative hypotheses:

- Purpose: The null hypothesis is tested for possible rejection, while the alternative hypothesis is what researchers aim to prove or support.
- Mathematical formulation: Null hypotheses typically use an equal sign (=), whereas alternative hypotheses use an inequality symbol (≠).
- Implications: If the null hypothesis is rejected, the alternative hypothesis is accepted, indicating a statistically significant effect or difference

3. What is the significance level in hypothesis testing, and why is it important ?

-> The significance level, denoted by α (alpha), is a crucial concept in hypothesis testing that determines the threshold for deciding whether the results are statistically significant. In simpler terms, it's the maximum probability of rejecting the null hypothesis when it's actually true, also known as a Type I error.

-> The significance level is important because :

- Determine the reliability of the results: By setting a significance level, it ensure that the observed effect is statistically significant and not just a result of chance.
- Avoid false positives: A low significance level (e.g., 0.01) reduces the risk of false positives, where to reject the null hypothesis when it's actually true.
- Balance Type I and Type II errors: The significance level helps to strike a balance between the risk of false positives (Type I error) and false negatives (Type II error).

4. What does a P-value represent in hypothesis testing ?

-> In hypothesis testing, the P-value represents the probability of obtaining a result that's at least as extreme as the one  observed, assuming the null hypothesis is true.

-> A small P-value (typically less than 0.05) indicates that  results are unlikely to occur by chance, suggesting that the null hypothesis is probably false. On the other hand, a large P-value (greater than 0.05) implies that the results could easily occur by chance, providing little evidence against the null hypothesis.

- P-value < 0.01: Highly statistically significant, strong evidence against the null hypothesis
- P-value < 0.05: Statistically significant, moderate evidence against the null hypothesis
- P-value > 0.05: Not statistically significant, little evidence against the null hypothesis

5.  How do you interpret the P-value in hypothesis testing ?

-> In hypothesis testing, a p-value represents the probability of obtaining results as extreme as, or more extreme than, the observed results, assuming the null hypothesis is true. A small p-value (typically less than 0.05) suggests strong evidence against the null hypothesis, leading to its rejection in favor of the alternative hypothesis.

-> Interpreting p-values:

- Small p-value (< 0.05):  Indicates that the observed results are unlikely to have occurred by chance alone, if the null hypothesis is true. This suggests strong evidence against the null hypothesis, and it's common practice to reject the null hypothesis in this case.

- Large p-value ( > 0.05):  Indicates that the observed results are likely to have occurred by chance alone, if the null hypothesis is true. This suggests weak evidence against the null hypothesis, and it's common practice to fail to reject the null hypothesis in this case.

9. How do you calculate the Z-score, and what does it represent in hypothesis testing ?

Z = (x̄ - μ) / (σ / √n)

-> In hypothesis testing, the Z-score is used to determine whether to reject the null hypothesis. If the Z-score is greater than the critical value or less than the negative critical value, you reject the null hypothesis. Otherwise, you fail to reject it.

-> The margin of error affects the confidence interval in that it determines the width of the interval. A larger margin of error results in a wider confidence interval, indicating more uncertainty in the estimate. On the other hand, a smaller margin of error leads to a narrower confidence interval, suggesting more precision in the estimate.

-> Several factors influence the margin of error, including:

- Sample Size: Larger sample sizes lead to smaller margins of error.
- Population Variability: More variability in the population results in a larger margin of error.
- Confidence Level: A higher confidence level, such as 99% instead of 95%, increases the margin of error .

20. What is an ANOVA test, and what are its assumptions ?

-> An ANOVA (Analysis of Variance) test is a statistical method used to compare the means of three or more groups to determine if there are any statistically significant differences among them. It's commonly used in various fields, such as biology, psychology, business, and social sciences, to analyze the differences between groups and identify if those differences are due to chance or if they indicate real effects.

-> Assumptions of ANOVA:

- Normal Distribution: The residuals should be normally distributed. However, ANOVA is relatively robust to minor violations of normality, especially when sample sizes are equal and greater than 10 .
- Homogeneity of Variances: The populations should have a common variance. A rule of thumb is that the ratio of the largest variance to the smallest variance should be less than 3 or 4 .
- Independence: All samples should be drawn independently of each other, and within each sample, the observations should be sampled randomly and independently of each other .
- Additivity: Factor effects should be additive, meaning that the effect of one factor does not depend on the level of another factor .
- No Outliers: There should be no significant outliers in the data, as they can affect the results of the ANOVA test .

21. What are the different types of ANOVA tests ?

-> There are several types of ANOVA tests, each used for different research designs and purposes.

- One-Way ANOVA: This type is used when one independent variable (factor) has three or more levels or groups. For example, comparing the mean test scores of students from different schools based on a single factor like teaching method .
- Two-Way ANOVA: This type involves two independent variables (factors) and is used when two main effects and interactions exist between them. For instance, comparing students' performance based on teaching method and gender .
- Factorial ANOVA: An extension of the two-way ANOVA that includes multiple factors with different levels, allowing for more complex designs and interactions.
- Repeated Measures ANOVA: This type is used when the same group of subjects is measured multiple times, such as before and after an intervention, to evaluate changes within the same subjects over time or conditions.
- MANOVA (Multivariate Analysis of Variance): An extension of ANOVA that handles multiple dependent variables, analyzing group differences across several outcomes.

22. What is the F-test, and how does it relate to hypothesis testing?

-> The F-test is a statistical test that compares variances, used in hypothesis testing to determine if the variances of two or more populations are significantly different, based on the F-statistic and the F-distribution.

-> How it relates to hypothesis testing:
- Null Hypothesis: The F-test typically tests the null hypothesis that the variances of the groups being compared are equal.
- Alternative Hypothesis: The alternative hypothesis is that the variances are not equal.
- F-statistic: The test calculates an F-statistic, which is the ratio of two variances (or mean squares).
-F-distribution: The calculated F-statistic is then compared to a theoretical distribution called the F-distribution.
- P-value: The p-value, derived from the F-distribution, indicates the probability of observing the obtained F-statistic (or a more extreme value) if the null hypothesis is true.
- Decision: If the p-value is below a pre-determined significance level (e.g., 0.05), the null hypothesis is rejected, suggesting that the variances are significantly different.

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and
interpret the results

```python
import numpy as np
from scipy import stats

sample_mean = 25.5
sample_std = 3.2
sample_size = 36

population_mean = 24.8

standard_error = sample_std / np.sqrt(sample_size)

z_score = (sample_mean - population_mean) / standard_error

p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Sample Size:", sample_size)
print("Known Population Mean:", population_mean)
print("Standard Error:", standard_error)
print("Z-score:", z_score)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis. The sample mean is significantly different from the known population mean.")
else:
    print("Fail to reject the null hypothesis. The sample mean is not significantly different from the known population mean.")
```

3.  Implement a one-sample Z-test using Python to compare the sample mean with the population mean

```python
def one_sample_z_test(sample, population_mean, population_std):

    sample_mean = np.mean(sample)

    standard_error = population_std / np.sqrt(len(sample))


    z_score = (sample_mean - population_mean) / standard_error


    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

    return z_score, p_value

np.random.seed(0)
sample = np.random.normal(loc=10, scale=2, size=100)
population_mean = 10
population_std = 2

z_score, p_value = one_sample_z_test(sample, population_mean, population_std)

print("Z-score:", z_score)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis. The sample mean is not significantly different from the population mean.")
```

4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot


```python
sample_mean = 25.5
population_mean = 24.8
population_std = 3.2

sample_size = 36

standard_error = population_std / np.sqrt(sample_size)

z_score = (sample_mean - population_mean) / standard_error

alpha = 0.05

critical_z_score = stats.norm.ppf(1 - alpha / 2)

print("Z-score:", z_score)
print("Critical Z-score:", critical_z_score)

z_values = np.linspace(-3, 3, 100)
plt.plot(z_values, stats.norm.pdf(z_values))
plt.axvline(x=-critical_z_score, color='r', linestyle='--')
plt.axvline(x=critical_z_score, color='r', linestyle='--')
plt.fill_between(z_values, stats.norm.pdf(z_values), where=(z_values < -critical_z_score) | (z_values > critical_z_score), color='r', alpha=0.2)
plt.title('Two-Tailed Z-Test Decision Region')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.show()
```

5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing



```python
def hypothesis_testing_errors(alpha, beta, mu0, mu1, sigma, sample_size):


    critical_region = stats.norm.ppf(1 - alpha, loc=mu0, scale=sigma / np.sqrt(sample_size))

    type1_error = alpha

    type2_error = stats.norm.cdf(critical_region, loc=mu1, scale=sigma / np.sqrt(sample_size))

    x = np.linspace(mu0 - 3 * sigma / np.sqrt(sample_size), mu1 + 3 * sigma / np.sqrt(sample_size), 100)
    plt.plot(x, stats.norm.pdf(x, loc=mu0, scale=sigma / np.sqrt(sample_size)), label='Null Hypothesis')
    plt.plot(x, stats.norm.pdf(x, loc=mu1, scale=sigma / np.sqrt(sample_size)), label='Alternative Hypothesis')
    plt.axvline(x=critical_region, color='r', linestyle='--', label='Critical Region')
    plt.fill_between(x, stats.norm.pdf(x, loc=mu0, scale=sigma / np.sqrt(sample_size)), where=(x > critical_region), color='r', alpha=0.2, label='Type 1 Error')
    plt.fill_between(x, stats.norm.pdf(x, loc=mu1, scale=sigma / np.sqrt(sample_size)), where=(x < critical_region), color='b', alpha=0.2, label='Type 2 Error')
    plt.legend()
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.title('Hypothesis Testing Errors')
    plt.show()

    return type1_error, type2_error

alpha = 0.05
beta = 0.2
mu0 = 0
mu1 = 1
sigma = 1
sample_size = 36

type1_error, type2_error = hypothesis_testing_errors(alpha, beta, mu0, mu1, sigma, sample_size)

print("Type 1 Error Rate:", type1_error)
print("Type 2 Error Rate:", type2_error)
```

6.Write a Python program to perform an independent T-test and interpret the results

```python
np.random.seed(0)
sample1 = np.random.normal(loc=25, scale=5, size=100)
sample2 = np.random.normal(loc=30, scale=5, size=100)

t_stat, p_value = stats.ttest_ind(sample1, sample2)

print("T-statistic:", t_stat)
print("P-value:", p_value)


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The means of the two samples are significantly different.")
else:
    print("Fail to reject the null hypothesis. The means of the two samples are not significantly different.")


plt.hist(sample1, alpha=0.5, label='Sample 1')
plt.hist(sample2, alpha=0.5, label='Sample 2')
plt.legend()
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Independent T-Test')
plt.show()

effect_size = (np.mean(sample1) - np.mean(sample2)) / np.sqrt(((len(sample1) - 1) * np.var(sample1) + (len(sample2) - 1) * np.var(sample2)) / (len(sample1) + len(sample2) - 2))
print("Effect size (Cohen's d):", effect_size)


if abs(effect_size) < 0.2:
    print("Small effect size")
elif abs(effect_size) < 0.5:
    print("Medium effect size")
else:
    print("Large effect size")
```

9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

```python
def calculate_confidence_interval(sample, confidence_level=0.95):

    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)

    standard_error = sample_std / np.sqrt(len(sample))

    degrees_of_freedom = len(sample) - 1
    critical_value = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom)

    margin_of_error = critical_value * standard_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

np.random.seed(0)
sample = np.random.normal(loc=25, scale=5, size=100)
confidence_level = 0.95

lower_bound, upper_bound = calculate_confidence_interval(sample, confidence_level)

print(f"Sample mean: {np.mean(sample)}")
print(f"Confidence interval ({confidence_level*100}%): ({lower_bound}, {upper_bound})")
```

11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process

```python
def bayes_theorem(prior_A, likelihood_B_given_A, likelihood_B_given_not_A, prior_not_A):
    numerator = prior_A * likelihood_B_given_A
    denominator = numerator + prior_not_A * likelihood_B_given_not_A
    posterior_A_given_B = numerator / denominator
    return posterior_A_given_B


prior_A = 0.5
likelihood_B_given_A = 0.9
likelihood_B_given_not_A = 0.1
prior_not_A = 1 - prior_A


posterior_A_given_B = bayes_theorem(prior_A, likelihood_B_given_A, likelihood_B_given_not_A, prior_not_A)

print("Posterior probability of having the disease given a positive test result:", posterior_A_given_B)
```

13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed
data

```python
def calculate_expected_frequencies(observed_data):
    total_rows = observed_data.sum(axis=1)
    total_cols = observed_data.sum(axis=0)
    grand_total = observed_data.sum()

    expected_frequencies = np.zeros(observed_data.shape)
    for i in range(observed_data.shape[0]):
        for j in range(observed_data.shape[1]):
            expected_frequencies[i, j] = (total_rows[i] * total_cols[j]) / grand_total

    return expected_frequencies

observed_data = np.array([[30, 20], [10, 40]])

expected_frequencies = calculate_expected_frequencies(observed_data)

print("Observed frequencies:")
print(observed_data)
print("\nExpected frequencies:")
print(expected_frequencies)
```

16. Implement an F-test using Python to compare the variances of two random samples

```python
np.random.seed(0)
sample1 = np.random.normal(loc=0, scale=1, size=100)
sample2 = np.random.normal(loc=0, scale=2, size=100)

var1 = np.var(sample1, ddof=1)
var2 = np.var(sample2, ddof=1)

f_stat, p_value = stats.f_oneway(sample1, sample2)

f_stat_alt = var1 / var2
p_value_alt = 2 * (1 - stats.f.cdf(f_stat_alt, len(sample1) - 1, len(sample2) - 1))

print("F-statistic:", f_stat)
print("P-value:", p_value)

if p_value < 0.05:
    print("Reject the null hypothesis. The variances are not equal.")
else:
    print("Fail to reject the null hypothesis. The variances are equal.")
```

19. Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA

```python
def check_anova_assumptions(data, groups):

    plt.figure(figsize=(8, 6))
    for i, group in enumerate(groups):
        plt.subplot(len(groups), 1, i+1)
        stats.probplot(data[group], dist="norm", plot=plt)
        plt.title(group)
    plt.show()

    correlations = data.corr()
    print("Correlation matrix:\n", correlations)

    levene_stat, levene_p_value = stats.levene(*[data[group] for group in groups])
    print("Levene's test statistic:", levene_stat)
    print("Levene's test p-value:", levene_p_value)
    if levene_p_value < 0.05:
        print("Reject the null hypothesis. Variances are not equal.")
    else:
        print("Fail to reject the null hypothesis. Variances are equal.")

np.random.seed(0)
data = pd.DataFrame({
    'Group1': np.random.normal(loc=10, scale=2, size=10),
    'Group2': np.random.normal(loc=12, scale=2, size=10),
    'Group3': np.random.normal(loc=11, scale=2, size=10)
})
check_anova_assumptions(data, ['Group1', 'Group2', 'Group3'])
```

```python
np.random.seed(0)

n = 100
mu = 0
sigma = 1
x = np.random.normal(mu, sigma, n)

s2 = np.var(x, ddof=1)

H0 = "The population variance is equal to 1."
H1 = "The population variance is not equal to 1."

alpha = 0.05

chi2_stat = (n - 1) * s2 / sigma**2

chi2_crit = stats.chi2.ppf(1 - alpha, n - 1)

if chi2_stat > chi2_crit:
    print("Reject the null hypothesis.")
    print("The population variance is not equal to 1.")
else:
    print("Fail to reject the null hypothesis.")
    print("The population variance is equal to 1.")

x = np.linspace(0, 20, 100)
y = stats.chi2.pdf(x, n - 1)
plt.plot(x, y)
plt.fill_between(x, y, where=(x > chi2_stat), color='red', alpha=0.5)
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density')
plt.title('Chi-square Distribution')
plt.show()
```

25. Write a Python script to perform a Z-test for comparing proportions between two datasets or groups

```python

from scipy import stats

n1 = 100
n2 = 100
p1 = 0.6
p2 = 0.4


p_pooled = (n1 * p1 + n2 * p2) / (n1 + n2)

se = np.sqrt(p_pooled * (1 - p_pooled) * (1 / n1 + 1 / n2))

z_score = (p1 - p2) / se

p_value = 2 * (1 - stats.norm.cdf(np.abs(z_score)))


print(f"Pooled proportion: {p_pooled:.4f}")
print(f"Standard error: {se:.4f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")


alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis. The proportions are significantly different.")
else:
    print("Fail to reject the null hypothesis. The proportions are not significantly different.")
```
